```python
import tensorflow as tf

graph_def_file = "Model/output_graph.pb"
input_arrays = ["numeric_input","codes_input"]
output_arrays = ["time_distributed_out/Reshape_1"]
input_shapes = {
    "numeric_input": [1,30,3],
    "codes_input": [1,1,1]
    }


converter = tf.lite.TFLiteConverter.from_frozen_graph(
  graph_def_file, input_arrays ,output_arrays,input_shapes)
tflite_model = converter.convert()
open("Model/converted_model.tflite", "wb").write(tflite_model)
```
